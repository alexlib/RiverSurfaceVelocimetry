```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
img = cv2.imread('1.png')
plt.imshow(img[:, :, ::-1])
plt.show()
```

```python
img_f = np.fft.fft2(img)
img_ac = np.fft.ifft2(np.abs(img_f) ** 2)
img_ac = np.fft.fftshift(img_ac)
img_ac_show = abs(img_ac)
img_ac_show = img_ac_show / img_ac_show.max() * 255

cv2.imshow('auto correlation', img_ac_show)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
